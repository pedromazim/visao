Explorando mudanças na resolução de imagens

```python
import numpy as np
from skimage import io
import matplotlib.pyplot as pt
```

```python
def make_gray(img, weights):

    '''
    converts an colored RGB image to grayscale

    img : image to be converted
    weights : list or array containing the relative weights of red, blue and green for conversion

    '''
    img_recol = np.zeros((img.shape[0],img.shape[1]))

    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
                img_recol[i,j] = weights[0]*img[i,j,0] + weights[1]*img[i,j,1] + weights[2]*img[i,j,2]

    return img_recol
```

```python
img = io.imread('https://github.com/pedromazim/visao/blob/main/lantana_flower.jpg?raw=true')
```

```python
fig = pt.figure(figsize=(5,5),dpi=100)
pt.imshow(img)
pt.axis('off')
```

```python
img_g = make_gray(img,[1/3,1/3,1/3])
```

```python
fig = pt.figure(figsize=(5,5),dpi=100)
pt.imshow(img_g, cmap='gray', vmin=0, vmax=255)
pt.axis('off')
```

```python
def rescale(img, scale_factor):

    '''
    changes image resolution by a given rescaling factor

    img : RGB or gray image in ndarray shape
    scale_factor : factor by which each image dimension is reduced

    '''

    if len(img.shape) == 3:
        img_rescaled = img[0:img.shape[0]:scale_factor,0:img.shape[1]:scale_factor,:]
    else:
        img_rescaled = img[0:img.shape[0]:scale_factor,0:img.shape[1]:scale_factor]
    return img_rescaled
```

```python
scale_factor=10
```

```python
fig = pt.figure(figsize=(10,10),dpi=100)
fig.tight_layout()
pt.imshow(rescale(img,scale_factor), cmap='gray', vmin=0, vmax=255)
pt.axis('off')
```

```python
fig = pt.figure(figsize=(10,10),dpi=100/scale_factor)
fig.tight_layout()
pt.imshow(rescale(img,scale_factor), cmap='gray', vmin=0, vmax=255)
pt.axis('off')
```
